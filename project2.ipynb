```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('all_data.csv')
```

```python
data = data.rename({"Life expectancy at birth (years)":"life"}, axis = "columns")
data = data.rename({"Country":"country"}, axis = "columns")
data = data.rename({"Year":"year"}, axis = "columns")
data = data.rename({"GDP":"gdp"}, axis = "columns")
data.head()
```

```python
plt.figure(figsize=(8,6))
sns.displot(data.gdp, rug = True, kde=False, color='red')
plt.xlabel("gdp");
```

```python
plt.figure(figsize=(8,6))
sns.lineplot(x=data.country, y=data.life, hue=data.year)
plt.legend(loc='center left', bbox_to_anchor=(1, 1), ncol=1)
plt.ylabel("life expectancy by year");
```

```python
graph_life = sns.FacetGrid(data, col="country", col_wrap=2, sharey = True)
graph_life = (graph_life.map(sns.regplot, "year", "life", color = 'green').set_axis_labels("year","life expentancy"))
```

```python
graph = sns.FacetGrid(data, col ='country', sharey = False)
graph.map(plt.hist, 'life', bins = 15, color ='orange')
plt.show()
```

```python
graph = sns.FacetGrid(data, col ="country",  hue ="country", col_wrap=3)
graph.map(plt.scatter, "year", "gdp").add_legend()
plt.show()
```
